```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

# Creating a Plot for the Proportion of 5 Stars and the Total number of Ratings a Movie Received

```python
movies = pd.read_csv('visualisation_five_stars/part-r-00000', sep='|', header=None)
```

```python
movies.columns = ["movieID", "title", "number_five_star_ratings", "number_ratings", "proportion_five_star_ratings"]
```

```python
number_ratings = np.array(movies["number_ratings"])
proportion_five_star_ratings = np.array(movies["proportion_five_star_ratings"])
```

```python
plt.scatter(number_ratings, proportion_five_star_ratings)
plt.title("The Proportion of Five Stars and Total Ratings per Movie ")
plt.xlabel("Number of Ratings")
plt.ylabel("Proportion of Ratings that were Five Stars")
plt.show()
```

# Creating a Plot to show the Total Number of Ratings given and the Average Rating given per User

```python
ratings = pd.read_csv('visualisation_users_average_rating/part-r-00000', sep='|', header=None)
```

```python
ratings.columns = ["userID", "total_ratings", "average_rating"]
```

```python
number_ratings = np.array(ratings["total_ratings"])
average_ratings = np.array(ratings["average_rating"])
```

```python
plt.scatter(number_ratings, average_ratings)
plt.title("The Number of Ratings compared to the Average Rating given per User")
plt.xlabel("Number of Ratings")
plt.ylabel("Average Rating")
plt.show()
```
